# **TIME COMPLEXITY CALCULATOR**

- The program will take an input code FILE (preferably coded in C with proper syntax in the sense every code block must start and end with curly braces including if and else or else if statements.)
- Calculates time taken to run for that program and also calculates the time complexity of the code written in the file.

<br>

### **WORKING**
- Read line by line. Start a timer.
- For each line read, 
  - CASE 1 : Finite/ Constant time = 1 => declaration, initialization, assignment, conditionals, basic arithmetic operations, indexing into arrays, ...
  - CASE 2: Loops => for, while, do ... while
  - *CASE 3: Recursion (not implementable in this way)* <br>
  We assign the time complexity for each line when read based on the above 3 cases
- After doing the above 2 steps for each line (when file has been fully read), do summation of the time complexities (not just normal sum because time complexity often is an expression in terms of 'n').
- End the timer and print the results.

```python
def split_code(code):
  code_stack = []
  bracket_stack = []
  splitted_code = []

  for line in code:
    if "{" in line:
      bracket_stack.append("{")
      code_stack.append("{")
    elif "}" in line:
      splitted_code.append(line.strip())
      tmp = bracket_stack.pop()
      if tmp == "{":
        current_line = code_stack.pop()
        while current_line != "{":
          splitted_code.append(current_line.strip())
          current_line = code_stack.pop()
        splitted_code.append(current_line.strip())
    else:
      code_stack.append(line.strip())

    print(code_stack," | ", bracket_stack, " | ", splitted_code)

  return splitted_code
```
